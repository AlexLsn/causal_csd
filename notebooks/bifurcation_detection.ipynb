# Automatic bif detection: Fig. S12

```python
t_len = 10000
```

```python
data = gendata.generate_time_series(t_len, sigma_Y = 0.01, seed = 1, model = 'base', return_noise=True, X_power = 1, bif_type='saddle-node')
```

```python
data_Y = data['data'][:, 1]
```

```python
proba_ts = shiftdetect.as_detect(data_Y, dt=None, lowwl=100, highwl=3000)
```

```python
threshold=-0.8
crossing_index = np.where(proba_ts <= threshold)[0][0]

fig = go.Figure()
fig.add_trace(go.Scatter(y=data_Y, mode='lines', name='Simulated tipping system', line=dict(color='black')))
fig.add_trace(go.Scatter(y=proba_ts, mode='lines', name='Proportion time series', line=dict(color='#A02C2C')))

fig.add_trace(go.Scatter(x=[0, len(proba_ts)], 
                             y=[threshold, threshold],
                             mode='lines', line=dict(color='grey', dash='dash'),
                             name='Threshold'))

fig.add_trace(go.Scatter(x=[crossing_index, crossing_index], 
                             y=[-1.5, 1.5],
                             mode='lines', line=dict(color='#DE8787', dash='dash', width=3),
                             name='Detected shift'))


fig.update_layout(height=450, width=700,
    xaxis=dict(range=[0, t_len],
    tickvals=[0, 2000, 4000, 6000, 8000, 10000], ticktext=['0', '2,000', '4,000', '6,000', '8,000', '10,000'], title='Time'),
    yaxis=dict(tickvals=[-1, 0, 1], ticktext=['-1', '0', '1']),
    showlegend=True,
    template='plotly_white',
    legend=dict(yanchor="top", y=0.99, xanchor="left", x=0.68),
    )

fig.show()
```
